# Data Preprocessing

```python
df = df.drop(columns=[
    'listing_id',
    'name',
    'district',
    'host_response_time',
    'host_response_rate',
    'host_acceptance_rate'
])
df.sample(5)
```

```python
# Replacing Numerical NaN
df['bedrooms'].fillna(df['bedrooms'].median(), inplace=True)

review_cols = [
    'review_scores_rating',
    'review_scores_accuracy',
    'review_scores_cleanliness',
    'review_scores_checkin',
    'review_scores_communication',
    'review_scores_location',
    'review_scores_value',
    'host_total_listings_count'
]

df[review_cols] = df[review_cols].fillna(0)
```

```python
# Replacing NaN categorical data
cat_cols = [
    'host_since',
    'host_location',
    'host_is_superhost',
    'host_has_profile_pic',
    'host_identity_verified'
]

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

# Data Splitting

```python
from sklearn.model_selection import train_test_split

X = df.drop('instant_bookable', axis=1)
y = df['instant_bookable']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.info()
```

# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder


low_card_cols = ['room_type', 'city']
prop_col = 'property_type'


top_k = 20
top_categories = X_train[prop_col].value_counts().nlargest(top_k).index


X_train[prop_col] = X_train[prop_col].where(
    X_train[prop_col].isin(top_categories), 'Other'
)

X_test[prop_col] = X_test[prop_col].where(
    X_test[prop_col].isin(top_categories), 'Other'
)


#  OneHotEncoder (low card)
ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
ohe.fit(X_train[low_card_cols])

X_train_low = ohe.transform(X_train[low_card_cols])
X_test_low  = ohe.transform(X_test[low_card_cols])

X_train_low_df = pd.DataFrame(
    X_train_low,
    columns=ohe.get_feature_names_out(low_card_cols),
    index=X_train.index
)

X_test_low_df = pd.DataFrame(
    X_test_low,
    columns=ohe.get_feature_names_out(low_card_cols),
    index=X_test.index
)


# OneHotEncoder (property_type)

ohe_prop = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
ohe_prop.fit(X_train[[prop_col]])

X_train_prop = ohe_prop.transform(X_train[[prop_col]])
X_test_prop  = ohe_prop.transform(X_test[[prop_col]])

X_train_prop_df = pd.DataFrame(
    X_train_prop,
    columns=ohe_prop.get_feature_names_out([prop_col]),
    index=X_train.index
)

X_test_prop_df = pd.DataFrame(
    X_test_prop,
    columns=ohe_prop.get_feature_names_out([prop_col]),
    index=X_test.index
)


X_train = X_train.drop(columns=low_card_cols + [prop_col])
X_test  = X_test.drop(columns=low_card_cols + [prop_col])

binary_cols = [
    'host_is_superhost',
    'host_has_profile_pic',
    'host_identity_verified'
]


mapping = {'t': 1, 'f': 0}


for col in binary_cols:
    X_train[col] = X_train[col].map(mapping)

for col in binary_cols:
    X_test[col] = X_test[col].map(mapping)

X_train = pd.concat([X_train, X_train_low_df, X_train_prop_df], axis=1)
X_test  = pd.concat([X_test, X_test_low_df, X_test_prop_df], axis=1)
```
