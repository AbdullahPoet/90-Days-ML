```python
from langchain_ollama import ChatOllama
```

```python
qwen = ChatOllama(
    model="qwen2.5:7b",
    temperature=0.7
)
```

```python
response = qwen.invoke("What is the capital of France?")
print(response.content)
```

```python
from langchain_core.messages import HumanMessage

response = qwen.invoke([
    HumanMessage(content="What is the capital of Canada?")
])

print(response.content)
```

```python
qwen.invoke([
    HumanMessage(content="What is the capital of Canada?"),
    HumanMessage(content="What is the largest city in Quebec?")
]).content
```
